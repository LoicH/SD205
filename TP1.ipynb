```python
import numpy as np
data = np.loadtxt('Motorcycledata.txt')
# print data
columns = ["ID", "Time", "Accel"]
ids, times, accels = data[...,0], data[...,1], data[...,2]

import matplotlib.pyplot as plt
%matplotlib inline
plt.scatter(times, accels)
plt.title("Acceleration of object")
plt.xlabel("Time (s)")
plt.ylabel("Acceleration")
```

```python
u_t, counts = np.unique(times, return_counts=True)

uniques = np.array(zip(u_t, counts)) #t, weight, mean
uniques = np.hstack([uniques, np.zeros((len(uniques),1))]) #adding a column of zeroes

for i in range(len(uniques)):
    uniques[i,2] = np.mean(accels[times == uniques[i,0]])
    
# print uniques
```

```python
def score(poly, x, y_true):
    """Return the squared errors between y_pred (i.e. poly(x)) and y_true"""
    return np.linalg.norm(np.poly1d(poly)(x)-y_true)**2

degres = range(1,17)
errors = []
interesting_degrees = [1, 2, 3, 5, 10, 15]

x = uniques[...,0]
y_true = uniques[...,2]
for degre in degres:
    poly = np.polyfit(x, y_true, degre)
    err = score(poly, x, y_true)
    errors.append(err)
#     print("Score with degre %s: %s" % (degre, err))
    if degre in interesting_degrees:
        plt.figure(degre)
        plt.title("With degree %s" % degre)
        plt.plot(x,np.poly1d(poly)(x))
        plt.scatter(x, y_true)
 
plt.figure(degre+1)
plt.scatter(degres, errors)
plt.title("Distance between prediction and true data")
plt.xlabel("Degree of polynome")
plt.ylabel("Distance")
```

```python
from scipy import interpolate

f = interpolate.interp1d(x, y_true, kind="cubic")
y_pred = f(x)

plt.plot(x,y_true, 'o', x, y_pred, '-')
```

## Création du vecteur *h*, de la matrice *Q* et *R*

```python
n = len(x)
# print(n)
h = []
for i in range(0,n-1):
    h.append(x[i+1] - x[i])
    
from scipy.sparse import csr_matrix

Q = csr_matrix((n, n-2))
for j in range(2,n-2):
    Q[j-1,j] = 1./h[j-1]
    Q[j,j]   = -1./h[j-1] - 1./h[j]
    Q[j+1, j]= 1./h[j]

    
```
